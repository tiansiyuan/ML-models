### Reproducing GPT-2

First download and tokenize the OpenWebText dataset. 

```python
!python data/openwebtext/prepare.py
```

To reproduce GPT-2 (124M) you'll want at least an 8X A100 40GB node and run:

```python
torchrun --standalone --nproc_per_node=8 train.py config/train_gpt2.py
```

```python
This will run for about 4 days using PyTorch Distributed Data Parallel (DDP) and go down to loss of ~2.85. 
```
